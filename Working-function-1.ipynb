### pairing identifiable name with identifiable rank, making a list

```python
nature_bear = {}
ff = open("new_formatted_bear.csv", "r", encoding = "ISO-8859-15")
ss = ff.readlines()[31:] #start reading from line 31
for line in ss:
    line = line.strip().strip('\n') #gets rid of all the empty lines
    if line:
        data = line.split(",") # ensures that index is not out of range
        nam = line.split(",")[5] #identifiable name
        ran = line.split(",")[6] #identifiable rank
        nature_bear[nam] = ran
ff.close()
```

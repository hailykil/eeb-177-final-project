````python
def makebearDict(filename):
    bearDict = {}
    fobj = open(filename, "r", encoding = "ISO-8859-15")
    all_records = fobj.readlines()[1:] 
    for line in all_records:
        record_elements = line.split(",")
        species = record_elements[1]
        interval = record_elements[3]
        bearDict[species] = interval  
    return bearDict
````

````python
bear = makebearDict("working_data.csv")
species = list(bear.keys())
species.sort()
bear
````

````python
def makebearDict(filename):
    bearDict = {}
    fobj = open(filename, "r", encoding = "ISO-8859-15")
    all_records = fobj.readlines()[1:] 
    for line in all_records:
        record_elements = line.split(",")
        interval = record_elements[0]
        species = record_elements[1]
        bearDict[interval] = species  
    return bearDict
````
